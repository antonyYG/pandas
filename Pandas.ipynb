Series


```python
#Crear una serie
serie =pd.Series([2,6,-1,8,9,2.3],index=['a','b','c','d','e','f'])
serie
```

```python
#Borrar series o filas
serie2 = serie.drop(['a','e'])
serie2
```

```python
#Seleccionar filaso series
serie[['b','d']]
```

```python
#asignacion
serie['b'] = 100
serie
```

```python
#operadores
serie4 = serie + serie2
serie4
```

```python
serie4.div(2)
```

```python
serie4.mul(2)
```

```python
serie.subtract(3)
```
